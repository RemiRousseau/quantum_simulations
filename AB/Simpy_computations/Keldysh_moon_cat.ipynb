```python
import sympy as sp
```

```python
aps, ap, am, ams, a, ad, ancilla = sp.symbols("a_+^* a_+ a_- a_-^* a a^\\dagger, ancilla")
display(aps, ap, am, ams, a, ad)
ac, acs, aq, aqs = sp.symbols("a_c a_c^* a_q a_q^*")
display(ac, acs, aq, aqs)
```

```python
def get_p_m(op):
    return op.subs({ad: aps, a: ap}), op.subs({ad: ams, a: am})

def get_lindb(H, L):
    Hp, Hm = get_p_m(H)
    Lp, Lm = get_p_m(L)
    Lps = Lp.replace(aps, ancilla).replace(ap, aps).replace(ancilla, ap)
    Lms = Lm.replace(ams, ancilla).replace(am, ams).replace(ancilla, am)
    Lcur = -sp.I * (Hp - Hm) + Lp*Lms - Lps*Lp/2 - Lms*Lm/2
    return sp.simplify(Lcur.expand())

def to_cl_q(op):
    op = op.replace(ap, (ac+aq)/sp.sqrt(2))
    op = op.replace(aps, (acs+aqs)/sp.sqrt(2))
    op = op.replace(am, (ac-aq)/sp.sqrt(2))
    op = op.replace(ams, (acs-aqs)/sp.sqrt(2))
    return sp.simplify(op.expand())
```

```python
w, k = sp.symbols("\\omega \\kappa")
k_2, al, lam = sp.symbols("\\kappa_2 \\alpha \\lambda")
H = 0*a
display(H)
L = (a**2 - al**2 + lam*ad*a) #sp.sqrt(k_2)*
display(L)
print('-------------')
Lindb = get_lindb(H, L)
display(Lindb)
Lindb = to_cl_q(Lindb)
display(Lindb)
print('-------------')
dt_alpha_c = sp.Derivative(Lindb, aqs).doit().simplify()
dt_alpha_q = sp.Derivative(Lindb, acs).doit().simplify()
display(dt_alpha_c)
display(dt_alpha_q)
print('-------------')
display(dt_alpha_c.subs({aq: 0, aqs:0}).simplify())
```
