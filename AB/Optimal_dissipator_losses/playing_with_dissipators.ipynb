```python
import dynamiqs as dq
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
N = 100
a, ad = dq.destroy(N), dq.create(N)

L2 = a @ a - 4 * dq.eye(N) + 0.1j * ad @ a @ a
```

```python
Hk = L2.dag() @ L2
eigH, eigsys = jnp.linalg.eigh(Hk.asdense().data)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
dq.plot.wigner(eigsys[:, 0].reshape((N, 1)), ax=ax[0], colorbar=False)
dq.plot.wigner(eigsys[:, 1].reshape((N, 1)), ax=ax[1], colorbar=False)
```
