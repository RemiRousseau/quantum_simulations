```python
import sympy as sp
```

# Simple case

```python
m, k, x, xd, l = sp.symbols("m k x \\dot{x} l")
c, coo, cx = sp.symbols("C C_\\infty C_x")
phi, phid, phibd = sp.symbols("\\varphi \\dot{\\varphi} \\dot{\\varphi_b}")
el = sp.symbols("E_L")
v0 = sp.symbols("V_0")
```

```python
U = k * x**2 / 2 + el * phi**2
T = m * xd**2 / 2 + c / 2 * phid**2 + cx / 2 * (phibd - phid) ** 2 + coo / 2 * phibd**2
display(T)
```

```python
p, Q, Qb = sp.symbols("p Q Q_b")
p_expr = sp.Derivative(T, xd).doit()
Q_expr = sp.Derivative(T, phid).doit().expand()
Qb_expr = sp.Derivative(T, phibd).doit().expand()
display(p_expr)
display(Q_expr)
display(Qb_expr)
```

```python
sol = sp.solve([Q_expr - Q, Qb_expr - Qb], (phid, phibd))
display(sol[phid])
display(sol[phibd])
```

```python
(
    c / 2 * sol[phid] ** 2
    + cx / 2 * (sol[phibd] - sol[phid]) ** 2
    + coo / 2 * sol[phibd] ** 2
).simplify()
```

```python
cg, v0 = sp.symbols("C_g V_0")

to_expand = Q**2 / 2 / (c + cx)
to_expand += Q * cx * v0 / (c + cx)
to_expand += c * cx * v0**2 / 2 / (c + cx)

display(to_expand)
exp = sp.series(to_expand.subs({cx: cg / (1 - x / l)}), x, 0, 3).simplify()
```

```python
for el in exp.args[-1].args:
    display((el * exp.args[0] * exp.args[1] * exp.args[2]).simplify())
```

# Complete case

```python
m, k, x, xd, l = sp.symbols("m k x \\dot{x} l")
c0, c, cp, cm, cb, cg = sp.symbols("C_0 C C_+ C_- C_b C_g")
phi0d, phi, phid, phimd = sp.symbols(
    "\\dot{\\varphi_0} \\varphi \\dot{\\varphi} \\dot{\\varphi_m}"
)
ej, el = sp.symbols("E_J E_L")
v0 = sp.symbols("V_0")
```

```python
U = k * x**2 / 2 - ej * sp.cos(phi) + el * phi**2
T = (
    m * xd**2 / 2
    + c0 / 2 * (phi0d + phid / 2) ** 2
    + c0 / 2 * (phi0d - phid / 2) ** 2
    + c / 2 * phid**2
    + cp / 2 * (phimd - phi0d - phid / 2) ** 2
    + cm / 2 * (phimd - phi0d + phid / 2) ** 2
    + cb / 2 * (v0 - phimd) ** 2
)
display(T)
```

```python
p, Qm, Q, Q0 = sp.symbols("p Q_m Q Q_0")
p_expr = sp.Derivative(T, xd).doit()
Qm_expr = sp.Derivative(T, phimd).doit()
Q_expr = sp.Derivative(T, phid).doit()
Q0_expr = sp.Derivative(T, phi0d).doit()
display(p_expr)
display(Qm_expr)
display(Q_expr)
display(Q0_expr)
```

```python
csigma, cdelta, cbar = sp.symbols("C_\\Sigma C_\\Delta \\bar{C}")
Q_ana = cbar * phid - cdelta * (phimd - phi0d)
Qm_ana = 2 * csigma * (phimd - phi0d) - cdelta * phid + cb * phimd - cb * v0
Q0_ana = -2 * (c0 + csigma) * (phimd - phi0d) + 2 * c0 * phimd + cdelta * phid
display(Q_ana)
display(Qm_ana)
display(Q0_ana)
sol = sp.solve([Q_ana - Q, Qm_ana - Qm, Q0_ana - Q0], (phid, phimd, phi0d))
display(sol[phid])
display(sol[phimd])
display(sol[phi0d])
phid_expr, phimd_expr, phi0d_expr = [
    sp.limit(sol[k], cb, sp.oo) for k in [phid, phimd, phi0d]
]
display(phid_expr)
display(phimd_expr)
display(phi0d_expr)
```

```python
csigma, cdelta = (cp + cm) / 2, (cp - cm) / 2
ctot = c0 / 2 + c + csigma / 2
beta = c0 / (c0 + csigma)

to_expand = Q**2 / 2 / ctot
to_expand = Q * beta * v0 * cdelta / ctot
to_expand = beta**2 * v0**2 * cdelta**2 / ctot
to_expand = beta * cdelta
to_expand = beta**2 * csigma * (1 + csigma / c0)

sp.series(to_expand.subs({cp: cg / (1 - x / l), cm: cg / (1 + x / l)}), x, 0, 4)
```
